```python
import pygame
import math

def make2DArray(rows, cols):
    arr = []
    for i in range(rows):
        arr.append([None] * cols)
    return arr

class Curve:
    def __init__(self):
        self.points = []
        self.x = 0
        self.y = 0

    def setX(self, x):
        self.x = x

    def setY(self, y):
        self.y = y

    def addPoint(self):
        self.points.append((self.x, self.y))

    def show(self, screen):
        if len(self.points) > 1:
            pygame.draw.lines(screen, (255, 255, 255), False, self.points, 1)

    def reset(self):
        self.points = []

pygame.init()
width, height = 400, 400
w = 40
cols = width // w - 1
rows = height // w - 1
curves = make2DArray(rows, cols)

for j in range(rows):
    for i in range(cols):
        curves[j][i] = Curve()

angle = 0
d = w - 0.2 * w
r = d / 2

screen = pygame.display.set_mode((width, height))
clock = pygame.time.Clock()
running = True
save_image = False

while running:
    screen.fill((0, 0, 0))

    for i in range(cols):
        cx = w + i * w + w // 2
        cy = w // 2
        pygame.draw.ellipse(screen, (255, 255, 255), (cx - r, cy - r, d, d), 1)
        x = r * math.cos(angle * (i + 1) - math.pi / 2)
        y = r * math.sin(angle * (i + 1) - math.pi / 2)
        pygame.draw.circle(screen, (255, 255, 255), (int(cx + x), int(cy + y)), 4)
        pygame.draw.line(screen, (255, 255, 255), (int(cx + x), 0), (int(cx + x), height), 1)
        
        for j in range(rows):
            curves[j][i].setX(cx + x)

    for j in range(rows):
        cx = w // 2
        cy = w + j * w + w // 2
        pygame.draw.ellipse(screen, (255, 255, 255), (cx - r, cy - r, d, d), 1)
        x = r * math.cos(angle * (j + 1) - math.pi / 2)
        y = r * math.sin(angle * (j + 1) - math.pi / 2)
        pygame.draw.circle(screen, (255, 255, 255), (int(cx + x), int(cy + y)), 4)
        pygame.draw.line(screen, (255, 255, 255), (0, int(cy + y)), (width, int(cy + y)), 1)
        
        for i in range(cols):
            curves[j][i].setY(cy + y)
            pygame.draw.circle(screen, (255, 255, 255), (int(curves[j][i].x), int(curves[j][i].y)), 4)

    for j in range(rows):
        for i in range(cols):
            curves[j][i].addPoint()
            curves[j][i].show(screen)

    angle -= 0.01

    if angle < -2 * math.pi:
        save_image = True
        for j in range(rows):
            for i in range(cols):
                curves[j][i].reset()
        angle = 0

    pygame.display.flip()
    clock.tick(60)

    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

#     if save_image:
#         pygame.image.save(screen, "desktop/lissajous_curve.png")
#         save_image = False

pygame.quit()
```
